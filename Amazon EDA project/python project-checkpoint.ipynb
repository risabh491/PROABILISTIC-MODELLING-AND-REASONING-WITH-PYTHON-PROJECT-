```python
import matplotlib.pyplot as plt

def extract_keywords(product_name):
    """Extracts keywords from a product name, handling potential numbers."""
    if isinstance(product_name, str):  # Check if it's a string
        keywords = product_name.lower().split()  # Split into words and lowercase
        keywords = [word for word in keywords if word.isalpha()]  # Keep only alphabetic words
    else:
        keywords = []  # Handle non-string values
    return keywords

# Apply the function to extract keywords
df["keywords"] = df["product_name"].apply(extract_keywords)

# Flatten the list of keywords
all_keywords = [keyword for keywords in df["keywords"] for keyword in keywords]

# Count keyword occurrences
keyword_counts = pd.Series(all_keywords).value_counts().head(10)

# Plot the top 10 keywords
plt.figure(figsize=(10,6))
keyword_counts.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Top 10 Most Popular Keywords", fontsize=14)
plt.xlabel("Keywords", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load dataset
df = pd.read_csv("amazon.csv")

# -----------------------------
# Data Cleaning
# -----------------------------
# Clean prices
df["discounted_price"] = df["discounted_price"].str.replace("‚Çπ", "").str.replace(",", "").astype(float)
df["actual_price"] = df["actual_price"].str.replace("‚Çπ", "").str.replace(",", "").astype(float)

# Clean discount percentage
df["discount_percentage"] = df["discount_percentage"].str.replace("%", "").astype(float)

# Convert ratings to float
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

# Convert rating_count to numeric
df["rating_count"] = df["rating_count"].str.replace(",", "").astype(float)

# -----------------------------
# Numerical Analysis
# -----------------------------
num_cols = ["discounted_price", "actual_price", "discount_percentage", "rating", "rating_count"]

for col in num_cols:
    data = df[col].dropna()
    print(f"\nüìä {col}")
    print(f"Mean: {data.mean():.2f}")
    print(f"Median: {data.median():.2f}")
    print(f"Trimmed Mean (10%): {stats.trim_mean(data, 0.1):.2f}")
    print(f"Range: {data.max() - data.min():.2f}")
    print(f"Variance: {data.var():.2f}")
    print(f"Standard Deviation: {data.std():.2f}")

    # Visualization
    plt.figure(figsize=(12,4))
    
    # Histogram
    plt.subplot(1,3,1)
    plt.hist(data, bins=30, edgecolor="black", alpha=0.7)
    plt.title(f"Histogram of {col}")

    # KDE Plot
    plt.subplot(1,3,2)
    sns.kdeplot(data, fill=True)
    plt.title(f"KDE Plot of {col}")

    # Boxplot
    plt.subplot(1,3,3)
    sns.boxplot(x=data)
    plt.title(f"Boxplot of {col}")

    plt.tight_layout()
    plt.show()

# -----------------------------
# Categorical Analysis
# -----------------------------
cat_cols = ["category"]

for col in cat_cols:
    print(f"\nüìä {col} Frequency Counts:")
    print(df[col].value_counts().head(10))  # top 10
    print(f"Mode: {df[col].mode()[0]}")

    plt.figure(figsize=(10,5))
    df[col].value_counts().head(10).plot(kind="bar")
    plt.title(f"Top 10 Categories in {col}")
    plt.ylabel("Frequency")
    plt.show()

# -----------------------------
# Correlation Analysis
# -----------------------------
plt.figure(figsize=(8,6))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# -----------------------------
# Grouped Boxplot / Barplot
# -----------------------------
# Example: Category vs Rating
plt.figure(figsize=(12,6))
sns.boxplot(x="category", y="rating", data=df)
plt.xticks(rotation=90)
plt.title("Rating Distribution across Categories")
plt.show()

# Example: Category vs Discount Percentage
plt.figure(figsize=(12,6))
sns.barplot(x="category", y="discount_percentage", data=df, estimator=np.mean, ci=None)
plt.xticks(rotation=90)
plt.title("Average Discount % by Category")
plt.show()

# -----------------------------
# Conclusion
# -----------------------------
print("\nüìå Conclusion / Key Insights:")
print("1. Actual prices are much higher than discounted prices, showing heavy discounting trends.")
print("2. The average product rating is fairly high (~4.0), but rating counts vary a lot across products.")
print("3. Categories differ significantly in discounts ‚Äì some consistently have higher % discounts.")
print("4. Discounted price and actual price are strongly correlated, as expected.")
print("5. Rating count is skewed: few products dominate with massive reviews while most have fewer.")
```
